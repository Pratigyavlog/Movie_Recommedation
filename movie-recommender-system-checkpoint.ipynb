```python
# import libraries
import numpy as np
import pandas as pd
```

```python
# upload dataset
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.head(1)
```

```python
credits.head(1)
```

```python
#merge the dataset
movies=movies.merge(credits,on='title')
movies.head(1)
```

```python
# genres id keywords title overview cast crew
movies=movies[['movie_id','title','keywords','genres','overview','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
```

```python
# retrieve only genere name from the tuple of genere list
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L

```

```python
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
#same for movies
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
# from cast series retireve the name of first three cast
def cast_name(text):
    L=[]
    counter=0
    for i in ast.literal_eval(text):
        if(counter!=3):
            L.append(i['name'])
            counter+=1
        else:
            break
    return L        
```

```python
movies['cast']=movies['cast'].apply(cast_name)
```

```python
movies.head()
```

```python
# for crew extract the persons for job title director
def extract_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L        
```

```python
movies['crew']=movies['crew'].apply(extract_director)
movies.head()
```

```python
# covert the overviw string into list
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# Remove the white spaces between the words
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags']=movies['overview']+movies['cast']+movies['genres']+movies['crew']+movies['keywords']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
new_df.head()
```
